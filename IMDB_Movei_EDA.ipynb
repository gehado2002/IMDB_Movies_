## üìÇ Dataset Overview


# üé¨ IMDb Top 1000 Movies and TV Shows Dataset - Data Overview

## üìñ Dataset Description
This dataset contains comprehensive information about the **Top 1000 Movies and TV Shows by IMDb Rating**, featuring detailed metadata for each title including ratings, cast, genre, runtime, and other key attributes.  

The dataset is ideal for:
- Exploratory Data Analysis (EDA)  
- Building movie recommendation systems  
- Understanding trends in highly-rated entertainment content  

It includes both **movies and TV shows** that have achieved the highest ratings on IMDb, providing a curated collection of critically acclaimed and audience-favorite content.
## Dataset Link  **https://www.kaggle.com/datasets/harshitshankhdhar/imdb-dataset-of-top-1000-movies-and-tv-shows**

---

## üìä Dataset Structure

- **Total Records:** 1,000 titles (movies and TV shows)  
- **Data Source:** IMDb (Internet Movie Database)  
- **Content:** Top-rated movies and TV shows  
- **File Format:** CSV  
- **Dataset Size:** ~200‚Äì300 KB  

---

## üìë Column Description

| Column Name    | Data Type | Description |
|----------------|-----------|-------------|
| `Poster_Link`  | String    | URL link to the movie/TV show poster image |
| `Series_Title` | String    | Title of the movie or TV show |
| `Released_Year`| Integer   | Year of release |
| `Certificate`  | String    | Age certification (e.g., PG, R, TV-MA) |
| `Runtime`      | String    | Duration of the movie/episode (e.g., `"142 min"`) |
| `Genre`        | String    | Genres associated with the title (comma-separated) |
| `IMDB_Rating`  | Float     | IMDb user rating (scale of 1‚Äì10) |
| `Overview`     | String    | Short plot summary or description |
| `Meta_score`   | Float     | Metacritic score (0‚Äì100), critic-based rating |
| `Director`     | String    | Director of the movie/TV show |
| `Star1`        | String    | Lead actor/actress |
| `Star2`        | String    | Second main actor/actress |
| `Star3`        | String    | Third main actor/actress |
| `Star4`        | String    | Fourth main actor/actress |
| `No_of_Votes`  | Integer   | Number of votes on IMDb |
| `Gross`        | Float     | Worldwide box office gross (in USD, may contain missing values) |

```python
pd.set_option('display.float_format', '{:.0f}'.format) #change the format in all of the EDA
df['Gross'] = df['Gross'].astype(str).str.replace(',', '', regex=True).astype(float)
df['No_of_Votes'] = df['No_of_Votes'].astype(int)
df['Released_Year'] = pd.to_numeric(df['Released_Year'], errors='coerce').astype('Int64')
df['Runtime'] = df['Runtime'].str.replace(' min', '').astype(float)
df['poster_img'] = df['Poster_Link'].apply(lambda x: f"<img src='{x}' width='100'>")
df.drop(columns=['Poster_Link'], inplace=True)
df.describe()
```

```python
#Detect nulls
null_counts = df.isnull().sum()
total_rows = len(df)
null_percentages = (null_counts / total_rows) * 100

null_summary = pd.DataFrame({
    'Column': null_counts.index,
    'Null_Count': null_counts.values,
    'Null_Percentage': null_percentages.values
})

null_columns_summary = null_summary[null_summary['Null_Count'] > 0]

if not null_columns_summary.empty:
    print("Columns with Null Values:")
    print(null_columns_summary)
else:
    print("No columns with null values found.")
```

- **Gross (169 nulls, 17%)**: üí∞ Fill using median imputation (less sensitive to outliers)





## üõ†Ô∏è Feature Engineering

```python
from itertools import combinations
from collections import Counter
# Rating categories
df['Rating_Category'] = pd.cut(df['IMDB_Rating'],
                                            bins=[0, 5.0, 7.0, 8.5, 10.0],
                                            labels=['Poor', 'Average', 'Good', 'Excellent'])

# Meta score categories
df['Meta_Category'] = pd.cut(df['Meta_score'],
                                          bins=[0, 40, 60, 80, 100],
                                          labels=['Poor', 'Mixed', 'Good', 'Universal'])

 # Rating gap (difference between IMDB and Meta score normalized)
df['Rating_Gap'] = df['IMDB_Rating'] - (df['Meta_score'] / 10)
df['Rating_Gap_Category'] = pd.cut(df['Rating_Gap'],
                                                 bins=[-np.inf, -1, 1, np.inf],
                                                 labels=['Critics_Favor', 'Balanced', 'Audience_Favor'])
# POPULARITY & ENGAGEMENT FEATURES
# Log transformation of votes
df['Log_Votes'] = np.log1p(df['No_of_Votes'])
# Popularity percentiles
df['Popularity_Percentile'] = pd.qcut(df['No_of_Votes'],
                                                   q=5,
                                                   labels=['Low', 'Below_Avg', 'Average', 'Above_Avg', 'High'])

# Engagement score (combines rating and popularity)
df['Engagement_Score'] = (df['IMDB_Rating'] *
                                       np.log1p(df['No_of_Votes']))

# Buzz ratio (votes per rating point)
df['Buzz_Ratio'] = df['No_of_Votes'] /df['IMDB_Rating']
# Success indicator (top 25% of engagement scores)
success_threshold = df['Engagement_Score'].quantile(0.75)
df['Is_Successful'] = (df['Engagement_Score'] >= success_threshold).astype(int)
current_year = 2025
df['Movie_Age'] = current_year - df['Released_Year']
# Era categorization
def categorize_era(year):
    if year < 1980:
        return 'Classic'
    elif year < 2000:
        return 'Modern'
    elif year < 2010:
        return 'Digital'
    else:
        return 'Contemporary'
df['Era'] =df['Released_Year'].apply(categorize_era)
# CERTIFICATE-BASED FEATURES
def group_certificate(cert):
    if cert in ['U', 'G']:  # U = Universal (India), G = General (US)
        return 'Family_Friendly'
    elif cert in ['UA', 'PG', 'GP', 'TV-PG', 'U/A']:
        return 'Parental_Guidance'
    elif cert in ['PG-13']:  # PG-13 = Parents Strongly Cautioned
        return 'Teen_Suitable'
    elif cert in ['A', 'R']:  # A = Adult (India), R = Restricted (US)
        return 'Adult_Only'
    elif cert in ['Passed', 'Approved']:  # Old US film ratings
        return 'Legacy_Approved'
    elif cert == 'Unrated':
        return 'Unrated'
    else:
        return 'Other'

df['Certificate_Group'] =df['Certificate'].apply(group_certificate)
# Create genre features
df['Genre_List'] = df['Genre'].str.split(',\s*')
df['Num_Genres'] = df['Genre_List'].apply(len)
df['Primary_Genre'] = df['Genre_List'].apply(
    lambda x: x[0] if isinstance(x, list) else None)

# Genre pairs analysis
pairs = Counter()
for genres in df['Genre_List'].dropna():
    if len(genres) >= 2:
        for combo in combinations(sorted(genres), 2):
            pairs[combo] += 1
genre_pairs = pd.DataFrame(pairs.most_common(15), columns=['Genre_Pair','Count'])

# Genre trios analysis
trios = Counter()
for genres in df['Genre_List'].dropna():
    if len(genres) >= 3:
        for combo in combinations(sorted(genres), 3):
            trios[combo] += 1
genre_trios = pd.DataFrame(trios.most_common(15), columns=['Genre_Trio','Count'])

# Explode genres for analysis
df_exploded = df.explode('Genre_List')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

numerical_cols = ['imdb_rating', 'meta_score', 'no_of_votes', 'gross', 'released_year', 'runtime']
colors = sns.color_palette('Set2', len(numerical_cols))

fig, ax = plt.subplots(figsize=(6,4))

def update(frame):
    ax.clear()
    col = numerical_cols[frame]
    sns.histplot(df[col], kde=True, ax=ax, color=colors[frame], edgecolor='black')
    ax.set_title(f'Distribution of {col.replace("_", " ").title()}')
    ax.set_xlabel(col.replace("_", " ").title())
    ax.set_ylabel('Frequency')

ani = FuncAnimation(fig, update, frames=len(numerical_cols), repeat=True, interval=1500)

# ÿßÿ∫ŸÑÿßŸÇ ÿßŸÑÿ±ÿ≥ŸÖ ÿßŸÑÿ£ÿ≥ÿßÿ≥Ÿä ÿπÿ¥ÿßŸÜ ŸÖÿß Ÿäÿ∏Ÿáÿ±ÿ¥ ÿ™ÿ≠ÿ™
plt.close(fig)

# ÿπÿ±ÿ∂ ÿßŸÑÿ£ŸÜŸäŸÖŸäÿ¥ŸÜ ŸÅŸÇÿ∑
HTML(ani.to_jshtml())
```

```python
categorical_cols = ['certificate', 'genre', 'director', 'star1', 'star2', 'star3', 'star4']
colors = ['skyblue', 'lightgreen', 'salmon', 'plum', 'orange', 'lightcoral', 'gold']

def create_animation():
    fig, ax = plt.subplots(figsize=(6,4))

    def update(frame):
        ax.clear()
        col = categorical_cols[frame]
        counts = df[col].value_counts().head(10)
        sns.barplot(x=counts.values, y=counts.index, ax=ax, palette=[colors[frame]]*len(counts))
        ax.set_title(f'Top 10 {col.replace("_", " ").title()}', fontsize=14)
        ax.set_xlabel('Count')
        ax.set_ylabel(col.replace("_", " ").title())

    ani = FuncAnimation(fig, update, frames=len(categorical_cols), repeat=True, interval=1500)
    plt.close(fig)  # prevents static image below
    return HTML(ani.to_jshtml())

create_animation()
```

```python
categorical_cols = ['certificate', 'genre', 'director', 'star1', 'star2', 'star3', 'star4']
colors = ['skyblue', 'lightgreen', 'salmon', 'plum', 'orange', 'lightcoral', 'gold']

def create_gross_animation():
    fig, ax = plt.subplots(figsize=(6,4))

    def update(frame):
        ax.clear()
        col = categorical_cols[frame]
        # ÿ≠ÿ≥ÿßÿ® ŸÖÿ™Ÿàÿ≥ÿ∑ gross ŸÑŸÉŸÑ ŸÅÿ¶ÿ©
        avg_gross = df.groupby(col)['gross'].mean().sort_values(ascending=False).head(10)
        sns.barplot(x=avg_gross.values, y=avg_gross.index, ax=ax, palette=[colors[frame]]*len(avg_gross))
        ax.set_title(f'Top 10 {col.replace("_", " ").title()} by Average Gross üí∞', fontsize=14)
        ax.set_xlabel('Average Gross ($)')
        ax.set_ylabel(col.replace("_", " ").title())

    ani = FuncAnimation(fig, update, frames=len(categorical_cols), repeat=True, interval=2000)
    plt.close(fig)  # ŸäŸÖŸÜÿπ ÿ∏ŸáŸàÿ± ÿµŸàÿ±ÿ© ÿ´ÿßÿ®ÿ™ÿ© ÿ£ÿ≥ŸÅŸÑ
    return HTML(ani.to_jshtml())

create_gross_animation()
```

```python
numerical_cols = ['imdb_rating', 'meta_score', 'no_of_votes', 'runtime', 'released_year']
colors = sns.color_palette('Set2', len(numerical_cols))

def create_bivariate_reg_animation():
    fig, ax = plt.subplots(figsize=(6,4))

    def update(frame):
        ax.clear()
        col = numerical_cols[frame]
        sns.regplot(
            x=col,
            y='gross',
            data=df,
            ax=ax,
            color=colors[frame],
            scatter_kws={'s':50, 'alpha':0.7}
        )
        ax.set_title(f'{col.replace("_", " ").title()} vs Gross üí∞ (Regression)', fontsize=16)
        ax.set_xlabel(col.replace("_", " ").title())
        ax.set_ylabel('Gross ($)')
        ax.grid(True)

    ani = FuncAnimation(fig, update, frames=len(numerical_cols), repeat=True, interval=2500)
    plt.close(fig)
    return HTML(ani.to_jshtml())

# ÿ™ÿ¥ÿ∫ŸäŸÑ ÿßŸÑÿ£ŸÜŸäŸÖŸäÿ¥ŸÜ
create_bivariate_reg_animation()
```

```python
import pandas as pd
import plotly.express as px
from plotly.subplots import make_subplots # Import make_subplots
import plotly.graph_objects as go # Import graph_objects for creating traces
# Calculate average gross and average votes for each star in each column
stars = ['star1', 'star2', 'star3', 'star4']
all_stars_data = []

for star_col in stars:
    # Calculate average gross and average votes for each star in the current column
    avg_data = df.groupby(star_col)[['gross', 'no_of_votes']].mean().reset_index()
    avg_data.rename(columns={star_col: 'Star', 'gross': 'Average_Gross', 'no_of_votes': 'Average_Votes'}, inplace=True)
    all_stars_data.append(avg_data)

# Combine the average data from all star columns
combined_stars_data = pd.concat(all_stars_data, ignore_index=True)

# Calculate the overall average gross and average votes for each unique star across all columns they appear in
overall_stars_data = combined_stars_data.groupby('Star')[['Average_Gross', 'Average_Votes']].mean().sort_values(by='Average_Gross', ascending=False)

# Define top and bottom N
top_n = 5
bottom_n = 5

# Create subplots: 1 row, 2 columns
fig = make_subplots(rows=1, cols=2,
                    subplot_titles=(f'Top {top_n} Stars by Overall Average Gross', f'Bottom {bottom_n} Stars by Overall Average Gross'))

# Create bar trace for top 5 stars
trace_top = go.Bar(x=overall_stars_data.head(top_n).index, y=overall_stars_data.head(top_n)['Average_Gross'])

# Create bar trace for bottom 5 stars
trace_bottom = go.Bar(x=overall_stars_data.tail(bottom_n).index, y=overall_stars_data.tail(bottom_n)['Average_Gross'])

# Add traces to subplots
fig.add_trace(trace_top, row=1, col=1)
fig.add_trace(trace_bottom, row=1, col=2)

# Update layout
fig.update_layout(height=600, width=1000, title_text="Overall Average Gross for Top and Bottom Stars")
fig.update_xaxes(title_text="Star", row=1, col=1)
fig.update_yaxes(title_text="Average Gross", row=1, col=1)
fig.update_xaxes(title_text="Star", row=1, col=2)
fig.update_yaxes(title_text="Average Gross", row=1, col=2)

fig.show()
```

```python
import plotly.express as px

# üîπ Create sequel column
def check_sequel(title):
    title = str(title).lower()
    keywords = ["part", "chapter", "episode", "ii", "iii", "iv", "v", "vi",
                "2", "3", "4", "5", "6", "7", "8", "9"]
    return any(k in title for k in keywords)

df['is_sequel'] = df['series_title'].apply(check_sequel)

# üîπ Group by sequel or not (Average Gross)
revenue_comparison = df.groupby('is_sequel')['gross'].mean().reset_index()
revenue_comparison['Category'] = revenue_comparison['is_sequel'].map({True: "üé¨ Sequels", False: "üé• Standalone"})

# üîπ Group by sequel or not (Count Percentage)
sequel_ratio = df['is_sequel'].value_counts(normalize=True) * 100
sequel_table = pd.DataFrame({
    "Category": ["üé¨ Sequels", "üé• Standalone"],
    "Percentage": [sequel_ratio.get(True, 0), sequel_ratio.get(False, 0)]
})

# ============== üìä Visualization ==============

# 1Ô∏è‚É£ Plotly Bar Chart (Interactive)
fig_bar = px.bar(
    revenue_comparison,
    x="Category",
    y="gross",
    color="Category",
    title="üí∞ Average Gross: Sequels vs Standalone Movies",
    text="gross",
    color_discrete_sequence=px.colors.qualitative.Set2
)
fig_bar.update_traces(texttemplate="$%{text:,.0f}", textposition="outside")
fig_bar.update_layout(xaxis_title="", yaxis_title="Average Gross ($)", showlegend=False)
fig_bar.show()

# 2Ô∏è‚É£ Plotly Pie Chart (Percentage of Movies)
fig_pie = px.pie(
    sequel_table,
    names="Category",
    values="Percentage",
    title="üé¨ Sequels vs üé• Standalone (Movie Share %)",
    color="Category",
    color_discrete_sequence=px.colors.qualitative.Set2,
    hole=0.4
)
fig_pie.update_traces(textinfo="label+percent", pull=[0.05, 0])
fig_pie.show()
```

## ü§ñ Recommendation System


```python
import ipywidgets as widgets
from IPython.display import display, HTML
import pandas as pd

# üîπ Year categories
def categorize_year(y):
    if y < 1980:
        return "üéûÔ∏è Old"
    elif y < 2000:
        return "üìÄ Classic"
    elif y < 2015:
        return "üé¨ Modern"
    else:
        return "üÜï Recent"

df['year_group'] = df['released_year'].apply(categorize_year)

# üîπ Genre categories
def categorize_genre(g):
    g = str(g).lower()
    if "comedy" in g:
        return "üòÇ Comedy"
    elif "drama" in g:
        return "üé≠ Drama"
    elif "action" in g:
        return "üí• Action"
    elif "horror" in g or "thriller" in g:
        return "üëª Horror/Thriller"
    elif "sci-fi" in g or "fantasy" in g:
        return "üåå Sci-Fi/Fantasy"
    elif "animation" in g or "family" in g:
        return "üë™ Family/Animation"
    elif "romance" in g:
        return "‚ù§Ô∏è Romance"
    elif "musical" in g:
        return "üé∂ Musical/Other"
    else:
        return "üé∂ Musical/Other"

df['genre_group'] = df['genre'].apply(categorize_genre)

# üîπ Certificate categories
def categorize_certificate(c):
    if pd.isna(c):
        return "‚ùì Unrated/Other"
    c = c.upper()
    if c in ["U", "UA", "PG", "G"]:
        return "üë®‚Äçüë©‚Äçüëß‚Äçüë¶ Family-Friendly"
    elif c in ["PG-13", "12A", "15"]:
        return "üéØ Teens"
    elif c in ["R", "A", "NC-17"]:
        return "üîû Adults"
    else:
        return "‚ùì Unrated/Other"

df['certificate_group'] = df['certificate'].apply(categorize_certificate)

# üîπ IMDb Rating Range
def categorize_rating(r):
    if r >= 8:
        return "‚≠ê 8+"
    elif r >= 7:
        return "‚≠ê 7+"
    elif r >= 6:
        return "‚≠ê 6+"
    else:
        return "‚≠ê All"

df['rating_group'] = df['imdb_rating'].apply(categorize_rating)

# üîπ Widgets
genre_dd = widgets.Dropdown(options=sorted(df['genre_group'].unique()), description='Genre:')
cert_dd = widgets.Dropdown(options=sorted(df['certificate_group'].unique()), description='Certificate:')
year_dd = widgets.Dropdown(options=sorted(df['year_group'].unique()), description='Year:')
rating_dd = widgets.Dropdown(options=["‚≠ê All", "‚≠ê 6+", "‚≠ê 7+", "‚≠ê 8+"], description='Rating:')

button = widgets.Button(description="üé¨ Get Recommendations", button_style="success")
output = widgets.Output()

# üîπ Recommendation function
def recommend(b):
    output.clear_output()
    with output:
        # Filter movies
        filtered = df[
            (df['genre_group'] == genre_dd.value) &
            (df['certificate_group'] == cert_dd.value) &
            (df['year_group'] == year_dd.value)
        ]

        # Apply rating filter
        if rating_dd.value == "‚≠ê 6+":
            filtered = filtered[filtered['imdb_rating'] >= 6]
        elif rating_dd.value == "‚≠ê 7+":
            filtered = filtered[filtered['imdb_rating'] >= 7]
        elif rating_dd.value == "‚≠ê 8+":
            filtered = filtered[filtered['imdb_rating'] >= 8]

        filtered = filtered.sort_values(by='imdb_rating', ascending=False).head(5)

        if filtered.empty:
            display(HTML("<h3 style='color:red;'>‚ö†Ô∏è No movies found with these filters!</h3>"))
        else:
            html = "<h2 style='color:#2c3e50;'>üçø Top Recommendations</h2>"
            for _, row in filtered.iterrows():
                html += f"""
                <div style='display:flex; align-items:flex-start; background:#ffffff;
                            border-radius:15px; margin:12px 0; padding:12px;
                            box-shadow:2px 2px 12px rgba(0,0,0,0.1);'>
                    <div style='margin-right:15px; flex-shrink:0;'>{row['poster_img']}</div>
                    <div>
                        <h3 style='margin:0; color:#e74c3c;'>{row['series_title']} ({row['released_year']})</h3>
                        <p style='margin:6px 0;'><b>üé≠ Genre:</b> {row['genre_group']}</p>
                        <p style='margin:6px 0;'><b>üîñ Certificate:</b> {row['certificate_group']}</p>
                        <p style='margin:6px 0;'><b>‚≠ê Rating:</b> {row['imdb_rating']} | <b>üéüÔ∏è Votes:</b> {row['no_of_votes']}</p>
                        <p style='margin:6px 0; color:#34495e;'><b>üìù Overview:</b> {row['overview']}</p>
                        <p style='margin:6px 0;'><b>üí∞ Gross:</b> ${row['gross']}</p>
                    </div>
                </div>
                """
            display(HTML(html))

button.on_click(recommend)

# üîπ Display UI
display(genre_dd, cert_dd, year_dd, rating_dd, button, output)
```
